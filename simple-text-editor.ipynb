```python
from guizero import App, Text, TextBox, PushButton, Box, Combo, CheckBox, Slider, MenuBar, info

def open_file():
    with open(app.select_file(title="Select file", folder=".", 
            filetypes=[["Text files", [".dat", "*.txt"]]], save=False), "r") as f:
        editor.value = f.read()
        
def save_as_file():
    global fname
    file_path = app.select_file(save=True)
    f = open(file_path, "w")
    f.write(editor.value)
    fname = f.name
    return fname

def save_file():
    with open(fname, "w") as f:
        f.write(editor.value)
    
def exit_app():
    exit_ans = app.yesno("Warning!", " Are you sure you want to exit?")
    if exit_ans == True:
        save_file()
        app.destroy()
    else:
        pass
    
def text_font():
    global font
    pref_controls.visible=True
    font = Combo(pref_controls, options=["verdana", "courier", "times new roman"], selected="times new roman", align="left",
                  command=change_font)
    return font

def change_font():
    editor.font = font.value
    
def text_size():
    global size
    pref_controls.visible=True
    size = Slider(pref_controls, start=8, end=42, align="left", command=change_text_size)
    return size
    
def change_text_size():
    editor.text_size = size.value
    editor.resize(1, 1)
    editor.resize("fill", "fill")
    
def text_colour():
    global colour
    pref_controls.visible=True
    colour = Combo(pref_controls, options=["black", "red", "blue", "green"], selected="black", align="left", command=change_colour)
    return colour

def change_colour():
    editor.text_color = colour.value

app = App(title="Text Editor", width=1260, height=720)

pref_controls = Box(app, align="top", width="fill", border=True, visible=False)

menubar = MenuBar(app, toplevel=["File", "Preferences"], options=[ 
           [["open", open_file], ["save", save_file], ["save as", save_as_file], ["exit", exit_app]], [["Fonts", text_font], ["Size", 
           text_size], ["Colour", text_colour]]])
file_controls = Box(app, align="top", width="fill", border=True, visible=False)
editor = TextBox(app, multiline=True, height="fill", width="fill")
editor.text_size=14

app.display()
```
