```python
import os, re
import numpy as np
from guizero import App, Box, Combo, PushButton, Text, TextBox, warn

app = App("Calculator", width=410, height=450)
result = Box(app, align="top", width="fill", height=150)
output = Text(result, width="fill", height=8, color="blue", size=30, align="right")
calc_input = TextBox(app, width="fill")
calc_input.text_size = 30
calc_keys = Box(app, layout="grid", width="fill", height="fill")
calc_keys.text_size=20
lg = np.log

def update(x):
    calc_input.value += x
    
def auto_correct():
    calc_input.value = output.value.split('\n')[-2].split('=')[0]
    
def evaluate():
    output.value += calc_input.value
    calc_input.value = re.sub("\u00B2", '**2', calc_input.value)
    calc_input.value = re.sub("\u00F7", '/', calc_input.value)
    calc_input.value = re.sub(r'\u221A([\w]+)|\u221A\((.+?)\)|\(\u221A(.+?)\)', r'np.sqrt(\1\2\3)', calc_input.value)
    calc_input.value = re.sub("(?<=\d)\(", '*(', calc_input.value)
    calc_input.value = re.sub("cos", 'np.cos', calc_input.value)
    calc_input.value = re.sub("sin", 'np.sin', calc_input.value)
    calc_input.value = re.sub("tan", 'np.tan', calc_input.value)
    calc_input.value = re.sub("cosh", 'np.cosh', calc_input.value)
    calc_input.value = re.sub("sinh", 'np.sinh', calc_input.value)
    calc_input.value = re.sub("tanh", 'np.tanh', calc_input.value)
    calc_input.value = re.sub("\u03C0", 'np.pi', calc_input.value)
    calc_input.value = re.sub("pi\(", 'pi*(', calc_input.value)
    calc_input.value = re.sub(r'([\w]+)e', r'\1*e', calc_input.value)
    calc_input.value = re.sub("e", '2.7182', calc_input.value)
    calc_input.value = re.sub("e^", 'np.exp', calc_input.value)
    calc_input.value = re.sub("ln", 'lg', calc_input.value)
    calc_input.value = re.sub("log", 'np.log10', calc_input.value)
    calc_input.value = re.sub(r'([\w]+)lg', r'\1*lg', calc_input.value)
    calc_input.value = re.sub(r'lg([\w]+)', r'lg(\1)', calc_input.value)
    calc_input.value = re.sub(r'log10([\w]+)', r'log10(\1)', calc_input.value)
    calc_input.value = re.sub(r'([\w]+)!|\((.+?)\)!', r'np.math.factorial(\1\2)', calc_input.value)
    calc_input.value = re.sub("\^", '**', calc_input.value)
    calc_input.value = re.sub(r'([\w]+)np', r'\1*np', calc_input.value)
    calc_input.value = re.sub(r'(.+?)\)np', r'\1)*np', calc_input.value)
    rslt = ("%.6f" % eval(calc_input.value))
    output.value += "=  "+str(rslt)+'\n'
    all_clear()
    
def all_clear():
    calc_input.value = ''
    
c1 = PushButton(calc_keys, text="7", width=3, height=1, grid=[0,0], command=lambda:update(c1.text))
c2 = PushButton(calc_keys, text="8", width=3, height=1, grid=[1,0], command=lambda:update(c2.text))
c3 = PushButton(calc_keys, text="9", width=3, height=1, grid=[2,0], command=lambda:update(c3.text))
c4 = PushButton(calc_keys, text=u"\u00F7", width=3, height=1, grid=[3,0], command=lambda:update(c4.text))
c5 = PushButton(calc_keys, text="AC", width=3, height=1, grid=[4,0], command=auto_correct)
c5.bg="yellow"
c6 = PushButton(calc_keys, text="C", width=3, height=1, grid=[5,0], command=all_clear)
c6.bg="red"
c7 = PushButton(calc_keys, text="4", width=3, height=1, grid=[0,1], command=lambda:update(c7.text))
c8 = PushButton(calc_keys, text="5", width=3, height=1, grid=[1,1], command=lambda:update(c8.text))
c9 = PushButton(calc_keys, text="6", width=3, height=1, grid=[2,1], command=lambda:update(c9.text))
c10 = PushButton(calc_keys, text="*", width=3, height=1, grid=[3,1], command=lambda:update(c10.text))
c11 = PushButton(calc_keys, text="(", width=3, height=1, grid=[4,1], command=lambda:update(c11.text))
c12 = PushButton(calc_keys, text=")", width=3, height=1, grid=[5,1], command=lambda:update(c12.text))
c13 = PushButton(calc_keys, text="1", width=3, height=1, grid=[0,2], command=lambda:update(c13.text))
c14 = PushButton(calc_keys, text="2", width=3, height=1, grid=[1,2], command=lambda:update(c14.text))
c15 = PushButton(calc_keys, text="3", width=3, height=1, grid=[2,2], command=lambda:update(c15.text))
c16 = PushButton(calc_keys, text="-", width=3, height=1, grid=[3,2], command=lambda:update(c16.text))
c17 = PushButton(calc_keys, text="x\u00B2", width=3, height=1, grid=[4,2], command=lambda:update("\u00B2"))
c18 = PushButton(calc_keys, text=u"\u221A", width=3, height=1, grid=[5,2], command=lambda:update(c18.text))
c19 = PushButton(calc_keys, text="0", width=3, height=1, grid=[0,3], command=lambda:update(c19.text))
c20 = PushButton(calc_keys, text=".", width=3, height=1, grid=[1,3], command=lambda:update(c20.text))
c21 = PushButton(calc_keys, text="%", width=3, height=1, grid=[2,3], command=lambda:update(c21.text))
c22 = PushButton(calc_keys, text="+", width=3, height=1, grid=[3,3], command=lambda:update(c22.text))
c23 = PushButton(calc_keys, text="x\u207F", width=3, height=1, grid=[4,3], command=lambda:update("^"))
c23.font = "Times New Roman"
c24 = PushButton(calc_keys, text="=", width=3, height=1, grid=[5,3], command=evaluate)
c24.bg="green"
c25 = PushButton(calc_keys, text="cos", width=3, height=1, grid=[6,0], command=lambda:update(c25.text))
c26 = PushButton(calc_keys, text="sin", width=3, height=1, grid=[7,0], command=lambda:update(c26.text))
c27 = PushButton(calc_keys, text="tan", width=3, height=1, grid=[8,0], command=lambda:update(c27.text))
c28 = PushButton(calc_keys, text="cosh", width=3, height=1, grid=[6,1], command=lambda:update(c28.text))
c29 = PushButton(calc_keys, text="sinh", width=3, height=1, grid=[7,1], command=lambda:update(c29.text))
c30 = PushButton(calc_keys, text="tanh", width=3, height=1, grid=[8,1], command=lambda:update(c30.text))
c31 = PushButton(calc_keys, text="e", width=3, height=1, grid=[6,2], command=lambda:update(c31.text))
c32 = PushButton(calc_keys, text=u"\u03C0", width=3, height=1, grid=[7,2], command=lambda:update(c32.text))
c33 = PushButton(calc_keys, text="!", width=3, height=1, grid=[8,2], command=lambda:update(c33.text))
c34 = PushButton(calc_keys, text="e\u207F", width=3, height=1, grid=[6,3], command=lambda:update("e^"))
c34.font="Times New Roman"
c35 = PushButton(calc_keys, text="ln", width=3, height=1, grid=[7,3], command=lambda:update(c35.text))
c36 = PushButton(calc_keys, text="log", width=3, height=1, grid=[8,3], command=lambda:update(c36.text))

app.display()
```
